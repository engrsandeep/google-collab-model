```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

companies = []
#  header and parsing pages method
def extract(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Safari/537.36'}
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    return soup.find_all('div', class_ = 'companies-results-company__info')
#    
def transform(articles):
    for item in articles:
        name = item.find('div', {'class': 'companies-results-company__title'}).find('a').string 
        if name:
          try:
              addresses = item.find('p', {'class': 'companies-results-company__address'}).text.strip()
              address = ' '.join(addresses.split())
          except:
              address = ''
          try:
              website = item.find('p', {'class': 'companies-results-company__site'}).find('a').text.strip()
          except:
              website = ''
          # try:
          #     tel = item.find('span', class_ = 'companies-results-company__phone').text.strip()
          # except:
          #     tel = ''

          company = {
              'name': name,
              'address': address,
              'website': website,
              # 'tel': tel
          }
          companies.append(company)
    return

def load():
    df = pd.DataFrame(companies)
    df.to_csv('bior_companies.csv', index=False)

letters = list(map(chr, range(ord('S'), ord('T')+1)))
for letter in letters:
    for x in range(1,19):
        print(f'Getting page {x}')
        print(f'Getting letter {letter}')
        articles = extract(f'https://www.cv-library.co.uk/companies/employers/{letter}?page={x}')
        transform(articles)
        time.sleep(5)

load()
print('Saved to CSV')
```
